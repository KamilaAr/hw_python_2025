### Задание №5: написать программу, которая с помощью цикла while определяет, сколько гласных/согласных из перечисленных в подстроке

```python
line = input()
vowels = 'ауеёэоыяию'
consonants = 'цкнгшщзхъфвпрлджчсмтбь'
count = 0
vc = 0
cc = 0
while count < len(line):
    if line[count] in vowels:
        vc += 1
    else:
        cc += 1
    count += 1

print(vc)
print(cc)
```

### Задание №6: написать программу, для проверки, явлеятся ли число квадратом

```python
num = int(input())
if num > 0:
    root = int(num ** 0.5)
    if root * root == num:
        print(f'Число {num} является квадратом')
    else:
        print(f'Число {num} не является квадратом')
else:
    print(f'Число отрицательное!')
```

### Задание №7: Доработать код задания 6, для проверки является ли число кубом

```python
num = int(input())
if num > 0:
    root = int(num ** 0.5)
    root3 = round(num ** (1/3))
    if root * root == num and root3 ** 3 == num:
        print(f'Число {num} является и квадратом, и кубом')
    elif root * root == num and root3 ** 3 != num:
        print(f'Число {num} является квадратом')
    elif root3 ** 3 == num and root * root != num:
        print(f'Число {num} является кубом')
    elif root3 ** 3 != num and root * root != num:
        print(f'Число {num} не является ни квадратом, ни кубом')
else:
    print(f'Число отрицательное!')
```

### Задание №8: Сделайте так, чтобы число секунд отображалось в виде "дни:часы:минуты:секунды"

```python
num = int(input())
seconds = 0
minutes = 0
hours = 0
days = 0
if num >= 86400:
    days += num // 86400
    num -= days*86400
if num >= 3600:
    hours += num // 3600
    num -= hours*3600
if num >= 60:
    minutes += num // 60
    num -= minutes*60
seconds = num 

print(f'{days}:{hours}:{minutes}:{seconds}')
```
