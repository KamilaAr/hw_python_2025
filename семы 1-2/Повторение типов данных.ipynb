## Условные операторы

```python
a = 5
b = 5
```

```python
a == b
```

```python
# проверка на равенство

if (a == b):
    print('equal')
```

```python
a = 5
b = 2
```

```python
# сравнение чисел

if (a > b):
    print('greater')
elif (a < b):
    print('less')
else:
    print('equal')
```

```python
# вложенные условные операторы

if (a > b):
    print('greater')
    if (a % 2 == 0):
        print('even')
    else:
        print('odd')
elif (a < b):
    print('less')
else:
    print('equal')
```

```python
a = 2
b = 4
```

```python
# сложные условия с и

if ((a > b) and (a % 2 == 0)):
    print('greater and even')
else:
    print('odd')
```

## Циклы


### while

```python
i = 0
k = 1
while (k <= 3) and (i <= 5):
    print(i)
    i += 1
    k += 1
```

### For

```python
for i in range(0, 6):
    print(i)
```
